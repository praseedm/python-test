```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
nifty_it = pd.read_csv('nifty_it.csv')
```

```python
close = pd.DataFrame(nifty_it.Close)
```

```python
def create_moving_avg(data):
    d = pd.DataFrame()
    for i in range(4,53,12):
        d[str(i)] = data.rolling(window=i).mean().fillna(0)
    return d    
```

```python
create_moving_avg(nifty_it.Close)
```
